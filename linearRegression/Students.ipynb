```python
import pandas as pd

# SETTINGS
TRAIN_FRACTION = .8

# Read CSV
students_df = pd.read_csv('student.csv', sep = ";")
students_df
```

```python
# Data Details
n = len(students_df.index)
train_size = n * TRAIN_FRACTION
test_size = n - train_size
train_size, test_size
```

```python
from sklearn.utils import shuffle
# Seprate train and test set
students_df = shuffle(students_df)
train_set = students_df.head(int(train_size))
test_set = students_df.tail(int(test_size))
```

```python
from sklearn.linear_model import LinearRegression

# I decided to only use G1 and G2
# More features can be added to the list below if desired
problem_features = ['G1', 'G2']
X = train_set[problem_features]
y = train_set['G3']
# Train using linear regression
reg = LinearRegression().fit(X, y)
reg.coef_
```

```python
# Test based on model
Xprime = test_set[['G1', 'G2']]
yprime = test_set['G3']

# Calculate Score
reg.score(X, y)
```
