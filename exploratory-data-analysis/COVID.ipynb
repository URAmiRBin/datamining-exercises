```python
import pandas as pd

# Read CSV from file
covid_df = pd.read_csv("covid.csv")
# Show all the rows
pd.set_option('display.max_rows', None)
covid_df
```

```python
print("Number of features : ", len(covid_df.columns))
print("Number of data : ", len(covid_df.index))
# Columns name
covid_df.columns
# Null count
covid_df.isnull().sum()
```

```python
# skipna for not taking account of NaN values
print("Birth Year Mean : ", covid_df.mean(axis = 0, skipna = True)['birth_year'])
print("Birth Year Std : ", covid_df.std(axis = 0, skipna = True)['birth_year'])
print("Birth Year Max : ", covid_df.max(axis = 0, skipna = True)['birth_year'])
```

```python
# Checks if there is any NaN
print(covid_df.isnull().values.any())
```

```python
# For NaN birth_year rows, use rounded mean birth
mean_birth = round(covid_df.mean(axis = 0, skipna = True)['birth_year'])
covid_df['birth_year'] = covid_df['birth_year'].fillna(mean_birth)
# For region and infection_resion use unknows
covid_df['region'] = covid_df['region'].fillna('Unknown')
covid_df['infection_reason'] = covid_df['infection_reason'].fillna('Unknown')
# For infected by use mean of infected by value
mean_infected_by = round(covid_df.mean(axis = 0, skipna = True)['infected_by'])
covid_df['infected_by'] = covid_df['infected_by'].fillna(mean_infected_by)
covid_df
```

```python
import matplotlib.pyplot as plt

# Histogram plot using birth_year
covid_df['birth_year'].plot(kind='hist')
plt.show()
```

```python
# Matrix plot
matrixplot = [[0 for x in range(176)] for y in range(2)]
# Sort the dataframe by sex to have a better view
# Left side is female and right side is male
sex_sorted = covid_df.sort_values('sex', ascending=True)
for i in range(176):
    # use different values for distinction
    value = 1000
    if sex_sorted['state'][i] == 'deceased' :
        value = 0
    elif sex_sorted['state'][i] == 'released' :
        value = 500
    # first row represents state and the second row represents infected_by
    matrixplot[0][i] = value
    # multiply by 2 for distinction
    matrixplot[1][i] = sex_sorted['infected_by'][i] * 2

plt.matshow(matrixplot)

plt.show()
```

```python
# sort dataframe by confirmed_date
sorted_date = covid_df.sort_values('confirmed_date', ascending=True)
# scatter using confirmed date and birth_year
plt.plot(sorted_date['confirmed_date'], sorted_date['birth_year'], '*')
plt.xticks(rotation='vertical')
```

```python
# Using the histogram, it can be concluded that there are some outliers
# For removing outliers, I cut of 5% quantiles
quan5 = covid_df[covid_df['birth_year'].between(covid_df['birth_year'].quantile(.025), covid_df['birth_year'].quantile(.975))] # without outliers

print("Removed ", len(covid_df) - len(quan5.index), " outliers")
quan5['birth_year'].plot(kind='hist')
plt.show()
```
